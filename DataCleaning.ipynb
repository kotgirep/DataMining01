Converstion of product-id from int to float type.

```python
product3['product-id'] = pd.to_numeric(product3['product-id'], downcast='float')
```

```python
plt.rcParams["figure.figsize"]=(10,7)
plt.plot(product3['product-id'],product3['product-size'])
plt.rcParams['agg.path.chunksize'] = 10000
```

Drawing a pie chart which is group by product-title and counting the distinct number of values.

```python
val1=product.groupby(by='product-title')['product-title'].count().values
labels1=product.groupby(by='product-title')['product-title'].count().index.values

plt.pie(x=val1, labels = labels1, radius = 1.6, autopct="%.1f%%")
plt.show()
```

Find out the correlation between columns product-id and product-size

```python
cor =product3.corr()
cor
```
